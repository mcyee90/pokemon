```python
pokemon_df = pd.read_csv("pokemon.csv")
pokemon_df.head()
```

```python
pokemon_df['total_attack'] = pokemon_df['attack'] + pokemon_df['sp_attack'] + pokemon_df['speed']
pokemon_df['total_defense'] = pokemon_df['defense'] + pokemon_df['sp_defense'] + pokemon_df['hp']
pokemon_df['color1'] = ''
pokemon_df['color2'] = ''

pokemon_df.head()
```

```python
pokemon_df = pokemon_df.drop([ 'attack',
                 'base_egg_steps',
                 'base_happiness',
                 'base_total',
                 'capture_rate',
                 'classfication',
                 'defense',
                 'experience_growth',
                 'height_m',
                 'hp',
                 'japanese_name',
                 'percentage_male',
                 'sp_attack',
                 'sp_defense',
                 'speed',
                 'weight_kg',
                 'generation',
                 'is_legendary'], axis = 1)
pokemon_df.head()
```

```python
pokemon_types = list(pokemon_df['type1'].unique())
print(pokemon_types)

colors = [['#35D15D','#030303'],['#FF0400','#F99735'],['#1500FD','#02EEF6'],['#38FF01','#FE00E0'],['#F7B1F5','#8F8E8F'],
['#E201DB','#E201DB'],['#FCF400','#040404'],['#5B3500','#A7A7A6'],['#FAE0F6','#010101'],['#F80F02','#5B2600'],['#EB00FC','#373737'],
['#A5A4A5','#D1C399'],['#030303','#FDFCFC'],['#A9F9FD','#D4D4D4'],['#080808','#E954F6'],['#060166','#A58F02'],['#080808','#E954F6'],
['#959594','#070707'],['#BCF8F9','#F2F1EE'],]
```

```python
c1 = []
c2 = []

for i, p in pokemon_df.iterrows():
    type_index = pokemon_types.index(p['type1'])
    color1 = colors[type_index][0]
    color2 = colors[type_index][1]
    c1.append(color1)
    c2.append(color2)
pokemon_df['color1'] = c1
pokemon_df['color2'] = c2
    
```

```python
pokemon_info = []
for i, p in pokemon_df.iterrows():
    if pd.isnull(p['type2']):
        type_text = '<br>Type:' + p['type1'].title()
    else:
        type_text = '<br>Types:' + p['type1'].title() +'/'+p['type2'].title()
    text = str(p['pokedex_number']) + '. ' + p['name'] + \
    "<br>Abilities: " + str(p['abilities']).replace('[','').replace("'",'').replace(']','') + type_text
    pokemon_info.append(text)
pokemon_df['info'] = pokemon_info
```

```python
pokemon_df.head()
```

```python
pokemon_df.to_csv('pokemon_analysis.csv')
```

```python
data = []
for types in pokemon_types:
    type_df = pokemon_df.loc[pokemon_df['type1'] == types]
    x = list(type_df['total_attack'])
#     print (x)
    y = list(type_df['total_defense'])
#     print(y)
    color1 = list(type_df['color1'])
    color2 = list(type_df['color2'])
#     print(color)
    text = list(type_df['info'])
#     print(text)
    name = types
    trace = go.Scatter(
        name = name,
        x = x,
        y = y,
        fillcolor = color1[0],
        text = text,
        mode = 'markers',
        marker=dict(size=10,
                   color=color1[0],
                   line = dict(color=color2[0],
                              width=1))
    )
    
    data.append(trace)  
print (data)
```

```python
layout = go.Layout(showlegend=True,legend=dict(x=1,y=1),title="Pokemon Chart")
```

```python
fig = go.Figure(
data=data,
layout=layout)

py.iplot(fig)
```
